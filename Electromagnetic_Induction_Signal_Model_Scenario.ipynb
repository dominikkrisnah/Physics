## Skenario Model

The case study presented will discuss more about the signal form (in general) from an electromagnetic induction which is influenced by magnetic field parameters (the magnetic field is basically the main influence on the formation of electromagnetic induction) which continues for a period of 10 seconds. The result is a visualization of the form of the signal which is a correlation or relationship between Induction (V/m) to the Magnetic Field (T) and to Time (s) which is processed and carried out using a model scenario in the form of a 2D model and a 3D model:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def electromagnetic_induction(time, magnetic_field):
    # function for calculating electromagnetic induction
    # parameters: time (time), magnetic field (magnetic_field)
    return magnetic_field * time

# example data of magnetic field and time
time = np.linspace(0, 10, 200)
magnetic_field = np.sin(time)

# calculate electromagnetic induction
induction = electromagnetic_induction(time, magnetic_field)

fig, ax = plt.subplots(figsize=[10,3.5])
ax.plot(time, induction, c='blue')
ax2=ax.twinx()
ax2.plot(time, magnetic_field, c='red')
ax.set_xlabel('Time [s]', c='black')
ax.set_ylabel('Induction\n[Volt/meter (V/m)]', c='blue')
ax2.set_ylabel('Magnetic field [T]', c='red')
ax.set_title("Electromagnetic Induction Signals\nCorrelation [2D]",c='black')
plt.show()
```

```python
fig, bgst = plt.subplots(figsize=[10,3.5])
bgst.plot(magnetic_field, induction, c='orange')
bgst.set_ylabel('Induction\n[Volt/meter (V/m)]', c='blue')
bgst.set_xlabel('Magnetic field [T]', c='red')
bgst.set_title("Magnetic Field vs Induction\nCorrelation [2D]",c='black')
plt.plot()
```

```python
# example data of magnetic field and time
time = np.linspace(0, 10, 200)
magnetic_field = np.sin(time)

# calculate electromagnetic induction
induction = electromagnetic_induction(time, magnetic_field)

fig = plt.figure(dpi=300,figsize=([5,5]))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(time, magnetic_field, induction, c='black')
ax.set_xlabel('Time [s]', c='green')
ax.set_ylabel('Magnetic field [T]', c='red')
ax.set_zlabel('Induction\n[Volt/meter (V/m)]', c='blue')
ax.set_title("Correlation between Induction, Magnetic field, and Time")
plt.show()
```
